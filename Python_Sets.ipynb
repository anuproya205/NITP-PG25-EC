# Sets in Python

```python
## Creating a set without values/ empty set
set_var=set()
print(set_var)
print(type(set_var))
```

```python
## Creating a set with values
fruits= {"","apple","mango","banana"}
print(fruits)
print(type(fruits))
```

```python
#adding an element
fruits.add("Orange")
print(fruits)
```

```python
# Difference
set1={1, 2, 3}
set2={2, 3, 4}
print(set1.difference(set2))## remove anything from set1 that is also in set2
print(set2.difference(set1))# remove anything from set2 that is also in set1
```
