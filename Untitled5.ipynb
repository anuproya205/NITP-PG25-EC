### loc(use row name) And iloc(use row numbers)

```python
import pandas as pd
import numpy as np
# Creating DataFrame(2D table) and store it in variable named df
df=pd.DataFrame(np.arange(0,20).reshape(10,2),index=['Row1','Row2','Row3','Row4','Row5','Row6','Row7','Row8','Row9','Row10'],columns=['Column1','Column2'])

df.loc['Row2'] #loc-> use row names

#df.iloc[1]  #iloc->use row

df.iloc[:,1:] #---brfore comma row (take all rows)after comma column(means start from col no 1 here c1 excluding)

df['Column1'] #--only print column1

df[['Column1','Column2']] #----multiple column



```
