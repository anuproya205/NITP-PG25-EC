#Matplotlib
### pythonlibrary used to visualize data andresult

```python
# 1. Import Matplotlib
import matplotlib.pyplot as plt
import numpy as np

np.pi  # 3.141

#Plot sine wave

#x=np.arange(start, stop, step)
#y=np.sin(x)
#plt.plot(x,y)
#plt.show

x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)  # takes sine of evry x value
plt.plot(x,y)
plt.show()
plt.subplot(2,2,1)
plt.plot(x,y,'r')  # Red line
plt.subplot(2,2,2)
plt.plot(x,y,'g')  # Green line
plt.subplot(2,2,3)
plt.plot(x,y,'bo')  # Blue line


```


#Bar Graph Plotting
