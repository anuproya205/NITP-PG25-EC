#NumPy-
####Numerical pyhton library used for (mathemetical calculation, arrays, matrix operation,signal processing ML, DL, Image processing)

```python
import numpy as np
#1D Array
my_list=[1,2,3,4,5]
arr= np.array(my_list)#---conver List into a NumPy array(used for mathematical calculations)
print(arr)
print(type(arr))
print(arr.shape)
#arr.reshape(7,8)

#2D Array
arr= np.array([[1,2], [3,4]])
print(arr)
print(arr.shape)
arr.reshape(4,1)

```

```python
import numpy as np
arr= np.array([1,2,3,4,5])
print(arr)
arr.shape #---dimesion of the array
```

```python
#Multidimentional array
my_lst1=[1,2,3,4]
my_lst2=[5,6,7,8]
my_lst3=[9,10,11,12]
arr= np.array([my_lst1,my_lst2,my_lst3]) #---or we can directly keep list inside np.array
print(arr)
arr.shape
arr.reshape(4,3)
arr[2]
#Slicing #---start from row 1 till the end and start from columns 3 till the end

arr[:,3:] #----select all rows, column 3
```

```python
#Selecting data from array (silicing)
#Creating a 3*5 Array
import numpy as np
arr=np.array([[1,2,3,4,5],[2,3,4,5,6],[9,7,6,8,9]])
print(arr)
arr.shape
arr[:,:]
arr[:,3:]
#arr[3:]
arr[0:2,0:2] #---take row0 and row1 and column 0 and column 1
```

```python
#Create numbers from 0 to 9
arr= np.arange(0,10)
print(arr)
```

```python
#Steps in Arange
arr=np.arange(0,10,step=2) #---start=0 , End=10 and jump=2
print(arr)
```

```python
# Create evenly spaced numbers --np.linspace()
np.linspace(0,10,5) #--- give me 5 number between 0 and 10
```
