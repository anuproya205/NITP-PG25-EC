##Dictionary -used to store information in the form key:value pair

### collection of data that stores information in the form of key:value pairs.

 Definition : A dictionary is a collection which is unordered , changeable , & indexed store many items together

```python
#1) Collection -- dictionary stores many items together

Student = { "Name" : "Anup",
            "Age"  : 25 ,
            "Branch" : "ece" }
print(Student)


```

```python
#2) Unordered -- data is not stored in a fixed order (it is organized by keys not by position)
```

```python
# 3)Changeable (Mutable)--value can change later
Student = { "Name" : "Anup",
            "Age"  : 25 ,
            "Branch" : "ece" }
Student["Age"] = 26
print(Student)

```

```python
#4) Indexed--we use key instead of number

Student = { "Name" : "Anup",
            "Age"  : 25 ,
            "Branch" : "ece" }

            #to get the name
print(Student["Name"])
```
