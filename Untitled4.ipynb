### Matrix Multiplication Rules

For two matrices `A` (dimensions `m x n`) and `B` (dimensions `n x p`), their matrix product `C = A @ B` will have dimensions `m x p`. The crucial part is that the number of columns in `A` (`n`) must equal the number of rows in `B` (`n`).

Your example `arr=np.array([[1,2,3,4,5],[2,3,4,5,6],[9,7,6,8,9]])` is a `3x5` matrix. If you want to multiply it with another matrix, the second matrix must have 5 rows.

```python
# Method1 -Matrix multiplication using @
import numpy as np
A=np.array([[1,2],[3,4]])
B=np.array([[5,6],[7,8]])
print(A)
print(B)
C=A@B
print(C)
#Method2--using np.dot
c=np.dot(A,B)
```
